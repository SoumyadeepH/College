```python
import matplotlib. pyplot as plt
import numpy as np
top10 = df.nlargest(10,'Average Attendance')
plt.bar(top10['Student Name'], top10['Average Attendance'])
plt.title("top 10 students by avg attendance: ")
plt.xticks(rotation = 90)
plt.ylabel("Avg Attendance")
plt.xlabel("Student Name")
plt.show()
```

```python
def category(x):
    if x >= 90:
        return 'Excellent'
    elif x >= 75:
        return 'Good'
    elif x >= 60:
        return 'Average'
    else:
        return 'Poor'
```

```python
df['Category'] = df['Average Attendance'].apply(category)
count = df['Category'].value_counts()
count.plot(kind='pie', autopct='%1.1f%%')
plt.title('Attendance categories')
plt.show()
```

```python
student= df[df['Student Name']=='PRATHAM GUPTA']
subject_attendance = student[
    [col for col in df.columns if 'Attendance %' in col and 'Average' not in col]
].values[0]

subjects= [
    'OS','DBMS','AI','Algo-II','Soft Skill','Design','Ethics'
]
plt.plot(subjects, subject_attendance, marker = 'o')
plt.title("Subject-wise attendance of PRATHAM GUPTA")
plt.grid(True)
plt.show()
```

```python
wholeclass = df['Average Attendance']
per = [0,40,55,70,85,100]
plt.hist( wholeclass, bins=per, edgecolor = 'red')
plt.title("Histogram Average Attendance")
plt.grid(axis = 'y', linestyle = "--", alpha = 0.7)
plt.show()
```
