```python
import pandas as pd
data=pd.read_csv('salary_data.csv')
print(data)
data.head()
```

```python
data.info()
data.describe()
```

```python
exp=data[data['YearsExperience'] >2.5]
print(exp)
```

```python
slr = data[data['Salary']>100000]
print(slr)
```

```python
minimum= data['Salary'].min()
print(minimum)
maximum= data['Salary'].max()
print(maximum)
avg= data['Salary'].mean()
print(avg)
```

```python
print ('Average Salary is: ',avg )
slavg=data[data['Salary']>avg]
print (slavg)
```

```python
correlation= data['YearsExperience'].corr(data['Salary'])
print(f"Correlation between Years of Experience and salary={correlation}")
```

```python
import matplotlib.pyplot as plt
plt.scatter(data ['YearsExperience'], data['Salary'], color=['red'])
plt.xlabel('YearsExperience')
plt.ylabel('Salary')
plt.title("Years of Experience and Salary")
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

x=data[['YearsExperience']]
y=data['Salary']

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 42)

model=LinearRegression()
model.fit(x_train,y_train)

print("intercept: ",model.intercept_)
print("coefficient: ", model.coef_[0])
```

```python
years=[[3]]
predict_salary=model.predict(years)
print(predict_salary)
```

```python
plt.scatter(data['YearsExperience'], data['Salary'], color='red')
plt.plot(data['YearsExperience'],model.predict(data[['YearsExperience']]), color='black')
plt.xlabel('YearsExperience')
plt.ylabel('Salary')
plt.title('Years of Experience and Salary')
plt.grid(True)
plt.show()
```
